#K-Nearest Neighbors

## Problem Set
We sell fish at the fish market.  
there are 2 type of fish.  
fish name are '도미', 빙어'
We have to classify the fish with length and weight.  

```
# library
import matplotlib.pyplot as plt # For the drawing graph
from sklearn.neighbors import KNeighborsClassifier # For the k-Nearest Neighbor algorithm
```

### Drow  scatter plot

```
# bream = '도미'
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]

# smelt = '빙어'
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

# Set lable: X=length, Y=weight
plt.xlabel('lenght')
plt.ylabel('weight')

# Set Data
plt.scatter(bream_length, bream_weight) # 도미
plt.scatter(smelt_length, smelt_weight) # 빙어

# Show Graph
plt.show()

# result of grapth: linear
```

Make data set (matrix)  
fish_data is X 

```
# bream size = 35, smelt size = 14
length = bream_length + smelt_length
weight = bream_weight + smelt_weight

# make n*2 matrix (n=size of brean + smelt = 49)
fish_data = [[l, w] for l, w in zip(length, weight)]

# same process
'''
fish_data = []
for l, w in zip(length, weight):
  ziped = []
  ziped.append(l)
  ziped.append(w)
  fish_data.append(ziped)
'''

print(fish_data)
```

Make target set  
fish_target is Y

```
bream_size = len(bream_length) # or len(bream_weight) result = 35
smelt_size = len(smelt_length) # or len(smelt_weight) result = 14
fish_size = bream_size + smelt_size

fish_target = [1] * bream_size + [0] * smelt_size 

print(fish_target)
```

Training with data set

```
# k-Nearest Neighbor
kn = KNeighborsClassifier() # declaration
kn.fit(fish_data, fish_target) # training with data set

# training result score 
# return: 0.5 ~ 1 
# Closer to 1, the better trained.
kn.score(fish_data, fish_target)
```

Make with new data

```
# new fish (Not classified)
new_fish_length = 30
new_fish_weight = 600
new_fish_data = [[new_fish_length, new_fish_weight]]

# Set lable: X=length, Y=weight
plt.xlabel('lenght')
plt.ylabel('weight')

# Set Data
plt.scatter(bream_length, bream_weight) # 도미
plt.scatter(smelt_length, smelt_weight) # 빙어
plt.scatter(new_fish_length, new_fish_weight) # ??

# Show Graph
plt.show()

# When you see the graph we can predict new fish is '도미'
```

Predict type of fish with new data

```
# When you see the graph we can predict new fish is '도미'
# Let's get the computer to do it. 

kn.predict(new_fish_data)
# result is 1 = '도미'
# Predicted success.
```

K-Nearest Neighbor option

```
# n_neighbors = number of nearest Neighbor
# '도미' = 35, '빙어' = 14
# If n_neighbors = 49, return value is allwayss '도미'
kn49 = KNeighborsClassifier(n_neighbors=49)

# training
kn49.fit(fish_data, fish_target)

# train_score is 0.714...
# because this model can only find '도미'
train_score = kn49.score(fish_data, fish_target)
print("Train Score: " + str(train_score))

# check way train score is 0.714...
# this value come from bream_size/fish_size. 
print("Check: " + str(bream_size/fish_size))
```
